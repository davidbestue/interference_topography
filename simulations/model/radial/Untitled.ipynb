```python
N=512
kappa_stim=100
theta = [float(range(0,N)[i])/N*2*pi for i in range(0,N)] 
p1=pi/2

stimulus2=zeros((N));
for i in range(0, N):
    stimulus2[i]=e**(kappa_stim*cos(theta[i] - p1)) / (2*pi*scipy.special.i0(kappa_stim)) 

    stimulus=stimulus2
stimulus=reshape(stimulus, (N,1))

plt.plot(stimulus)
```
