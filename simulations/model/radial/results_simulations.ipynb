```python
dfd= pd.read_excel('C:\\Users\\David\\Desktop\\IDIBAPS\\simulations_radial\\results_simul_.xlsx')
dfd['time']='delay'

dfp = pd.read_excel('C:\\Users\\David\\Desktop\\IDIBAPS\\simulations_radial\\results_simul_d0.xlsx')
dfp['time']='perception'

df=pd.concat([dfp, dfd])
```

```python
c_d0='grey'
c_d3 = '#91cf60'

sns.lineplot(x="position", y="interference", hue='time', hue_order=['perception', 'delay'],
             ci=95, palette=[c_d0, c_d3], data=df)

plt.gca().spines['right'].set_visible(False)  # aesthetics                                                                              
plt.gca().spines['top'].set_visible(False)                                                                                  
plt.gca().get_xaxis().tick_bottom()                                                                                         
plt.gca().get_yaxis().tick_left()
plt.gca().tick_params(direction='in') #direction
plt.ylabel(r'attraction to fixation ($^\circ$)', fontsize=25);
plt.xlabel(r'eccentricity ($^\circ$)', fontsize=25);
plt.ylim(-2,12);
plt.yticks([0,4,8,12]);
plt.plot([60, 300], [0,0], 'k--', linewidth=1)

l =plt.legend(loc=2, frameon=False, prop={'size': 18})

hue_order=['', 'delay 0', 'delay 3']
palette=['black', c_d0, c_d3]

for i_h, h_idx in enumerate(hue_order):
    l.get_texts()[i_h].set_text(h_idx)
    l.legendHandles[i_h].set_visible(False);
    l.get_texts()[i_h].set_color(palette[i_h]);


                   
```
