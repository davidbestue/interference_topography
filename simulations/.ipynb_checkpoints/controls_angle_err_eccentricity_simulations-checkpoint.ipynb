```python

Df=pd.read_csv('C:\\Users\\David\\Dropbox\\IDIBAPS\\my code\\Project 1 (master thesis)\\Modeling angular dimension\\FENS_18\\GEE_simulations_2.csv')
df  = Df.loc[Df['bias']<20]


#Liniar model
result = smf.ols(formula="bias ~  GEE", data=df).fit()
print(result.summary())

#Plot
plt.figure()
h = seaborn.factorplot('GEE', 'bias', data=df,ci=None, color='k')
plt.gca().set_title('BUMP bias with GEE', fontname='sans-serif', fontsize=15, weight ='bold')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().get_xaxis().tick_bottom()
plt.gca().get_yaxis().tick_left()
#plt.title('BUMP bias with GEE', fontsize=10)
plt.xlabel('GEE', fontname='sans-serif', weight ='bold' )
plt.ylabel('Bias (deg)', fontname='sans-serif', weight ='demibold')
h.set(xticklabels=[ 0.02, '', 0.021, '', 0.022, '', 0.023])

cis=[]
for i, Gee in enumerate(df.GEE.unique()):
    C = ci(df.groupby('GEE').get_group(Gee).bias, statfunction=numpy.mean)
    C = [round(C[n], 2) for n in[0,1]]
    cis.append(C)


plt.fill_between(list(range(0,len(df.GEE.unique()))),array(cis)[:,0],array(cis)[:,1],color="green",alpha=0.3)
plt.show(block=False)


#Plot
plt.figure()
h = seaborn.factorplot('GEE', 'Decoded_ang', estimator=numpy.std, data=df,ci=None, color='k')
plt.gca().set_title('Decoded angle std with GEE', fontname='sans-serif', fontsize=15, weight ='bold')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().get_xaxis().tick_bottom()
plt.gca().get_yaxis().tick_left()
#plt.title('BUMP bias with GEE', fontsize=10)
plt.xlabel('GEE', fontname='sans-serif', weight ='bold' )
plt.ylabel('Bias (deg)', fontname='sans-serif', weight ='demibold')
h.set(xticklabels=[ 0.02, '', 0.021, '', 0.022, '', 0.023])

cis=[]
for i, Gee in enumerate(df.GEE.unique()):
    C = ci(df.groupby('GEE').get_group(Gee).Decoded_ang, statfunction=numpy.std)
    C = [round(C[n], 2) for n in[0,1]]
    cis.append(C)


plt.fill_between(list(range(0,len(df.GEE.unique()))),array(cis)[:,0],array(cis)[:,1],color="green",alpha=0.3)
plt.show(block=False)
```
