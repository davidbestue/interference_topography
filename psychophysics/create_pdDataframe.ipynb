```python
#Function subject-code

def subject_number(code_subj):
    if code_subj == 'AnaHL':
        index=1
    elif code_subj == 'carlesmoya':
        index=2
    elif code_subj == 'cristinafernandez':
        index=3
    elif code_subj == 'cristinaxufre':
        index=4
    elif code_subj == 'davidnew':
        index=5
    elif code_subj == 'eduard':
        index=6        
    elif code_subj == 'elenamast':
        index=7 
    elif code_subj == 'evatrist':
        index=8
    elif code_subj == 'ferran':
        index=9
    elif code_subj == 'jordisoler':
        index=10
    elif code_subj == 'juliap':
        index=11
    elif code_subj == 'Laura':
        index=12
    elif code_subj == 'MarcLab':
        index=13
    elif code_subj == 'marcpl':
        index=14
    elif code_subj == 'mariacast':
        index=15
    elif code_subj == 'natalia':
        index=16
    elif code_subj == 'onacas':
        index=17
    elif code_subj == 'Raquel':
        index=18
    else:
        index=99999
    
    
    return index
```

```python
#Create data file

Dfs=[]

for file_to_use in os.listdir():
    print(file_to_use)
    df1 = pd.read_table(file_to_use)
    ##
    df1 = df1[df1['choiceAngle']<90] #max error allowed
    df1 = df1[df1['choiceAngle']>-90] #max error allowed
    ##
    df1['A_R'] = df1['angle1']-df1['choiceAngle']
    df1['R_R'] = df1['ring1']-df1['choiceR']
    df = df1[['ring1', 'ring2', 'delay', 'type', 'angle1', 'angle2', 'choiceAngle', 'choiceR', 'A_R', 'R_R']]
    df.columns=['r_T', 'r_NT', 'delay', 'type', 'A_T', 'A_NT', 'A_err', 'r_err', 'A_R', 'r_R']
    df = df[df['A_err']<9999] #take off nans
    #subject and session
    subj = file_to_use.split('_')[0]
    sess = file_to_use.split('_')[1][-1]
    df['subject']=subject_number(subj)
    df['session']=int(sess)
    
    Dfs.append(df)

    
data=pd.concat(Dfs, ignore_index=True)
data
```

```python
#Correct possible mistakes
data.loc[data['A_R']>360, 'A_R'] = data.loc[data['A_R']>360, 'A_R'] - 360
```

```python
def circ_dist(a1,a2):
    ## Returns the minimal distance in angles between to angles 
    op1=abs(a2-a1)
    angs=[a1,a2]
    op2=min(angs)+(360-max(angs))
    options=[op1,op2]
    return min(options)


def Interference_effects(target, response, reference):
    from numpy import array
    #input list of target, list of responses and list of references
    #Error_interference; positive for attraction and negative for repulsion
    #######
    #Decimals to get
    decimals=3
    ####
    interferences=[]
    for i in range(0, len(target)):
        angle_err_abs=abs(circ_dist(target[i],response[i]))
        if circ_dist(array(response)[i], array(reference)[i])<=circ_dist(array(target)[i], array(reference)[i]):
            Err_interference=round( angle_err_abs, decimals) 
        else:
            Err_interference=round( -angle_err_abs, decimals)

        interferences.append(Err_interference)
    
    return interferences



```

```python
data['Int_A_err'] = Interference_effects(list(data['A_T']), list(data['A_R']), list(data['A_NT']))
#In some conditions, the interference effect in angle makes no sense (controls and radial condition)
data.loc[data['type']==0, 'Int_A_err']=np.nan #controls
data.loc[data['type']==3, 'Int_A_err']=np.nan #in
data.loc[data['type']==4, 'Int_A_err']=np.nan #out
```

```python
data['Int_r_err'] = Interference_effects(list(data['r_T']), list(data['r_R']), list(data['r_NT']))
#In some conditions, the interference effect in angle makes no sense (controls and radial condition)
data.loc[data['type']==0, 'Int_r_err']=np.nan #controls
data.loc[data['type']==1, 'Int_r_err']=np.nan #in
data.loc[data['type']==2, 'Int_r_err']=np.nan #out
```

```python
#replace the -1 for Nans
data.loc[data['r_NT']==-1, 'r_NT']=np.nan
data.loc[data['A_NT']==-1, 'A_NT']=np.nan
```
